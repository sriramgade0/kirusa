```python
recipe_text = """
### **Recipe Template**

#### 1. **Dish Name**  
> Idli

#### 2. **Description**  
> Idli is a soft, fluffy steamed rice cake that is a staple breakfast dish in South India. It is healthy, light, and pairs perfectly with chutneys and sambar.

---

#### 3. **Ingredients**  
- **For Batter:**
  - 1 cup rice
  - 1/4 cup urad dal
  - 1/2 tsp fenugreek seeds
  - Water (as required)
- **For Steaming:**
  - Oil (for greasing the idli molds)

---

#### 4. **Preparation Time**  
- **Preparation Time:** 6-8 hours (for soaking and fermenting)  
- **Cooking Time:** 10-15 minutes  
- **Total Time:** ~8 hours 15 minutes  

---

#### 5. **How to Prepare**  
1. Soak Ingredients: Wash the rice and urad dal separately. Soak them along with fenugreek seeds for 6 hours.
2. Grind Batter: Grind the soaked urad dal into a smooth paste. Then grind the rice into a slightly coarse batter. Combine both, add salt, and mix well.
3. Ferment Batter: Allow the batter to ferment overnight or for 8 hours in a warm place. The batter should rise and become light.
4. Steam Idlis: Grease idli molds with oil, pour the batter, and steam for 10-15 minutes in an idli steamer.
5. Serve: Serve hot with coconut chutney and sambar.

---

#### 6. **Tips**  
- Use leftover idlis to make masala idli or fry them for a snack.  
- Adjust fermentation time based on the climate. Warm weather speeds up fermentation.

---

#### 7. **Serving Suggestions**  
- Serve with sambar, coconut chutney, or tomato chutney.  
- Add a dollop of ghee on top for extra flavor.

---

#### 8. **Nutritional Information**  
- **Calories:** 120 kcal  
- **Protein:** 4 g  
- **Carbohydrates:** 22 g  
- **Fat:** 0.5 g  
"""
```

```python
import numpy as np
import pandas as pd
import re
import os

# Define the CSV file path
csv_file = "recipes.csv"

# Load existing data if the CSV exists; otherwise, create an empty DataFrame
if os.path.exists(csv_file):
    df = pd.read_csv(csv_file)
else:
    df = pd.DataFrame(columns=["Name", "Desc (Preparation)", "Photo", "Region", "Time of Eating (B)", 
                               "Time of Eating (L)", "Time of Eating (D)", "Food Type", "Ratings",
                               "Cholesterol", "Sodium", "Saturated Fat", "Sugar", "Type", 
                               "Food Type (Veg/Non-Veg/Vegan)", "Category (Breakfast/Lunch/Dinner/Snack/Beverages)", 
                               "Category (Diabetic/BP)", "Category (Muscle Gain)"])

def add_recipe(recipe_text, patterns):
    global df  # Ensure df is updated globally
    
    # Extract data from recipe_text
    extracted_data = {}
    for key, pattern in patterns.items():
        match = re.search(pattern, recipe_text, re.DOTALL)
        extracted_data[key] = match.group(1).strip() if match else np.nan
    
    # Add additional fixed fields
    extracted_data.update({
        "Desc (Preparation)": extracted_data.get("How to Prepare"),
        "Photo": np.nan,
        "Region": "South India",
        "Time of Eating (B)": True,
        "Time of Eating (L)": False,
        "Time of Eating (D)": False,
        "Food Type": "Healthy, Gluten-Free",
        "Ratings": 5,
        "Cholesterol": 0,
        "Sodium": 0,
        "Saturated Fat": 0,
        "Sugar": 0,
        "Type": "Steamed Dish",
        "Food Type (Veg/Non-Veg/Vegan)": "Veg",
        "Category (Breakfast/Lunch/Dinner/Snack/Beverages)": "Breakfast",
        "Category (Diabetic/BP)": "Diabetic Friendly",
        "Category (Muscle Gain)": np.nan
    })
    
    # Update DataFrame
    name_to_update = extracted_data.get("Name")

    if name_to_update in df["Name"].values:
        for col, value in extracted_data.items():
            df.loc[df["Name"] == name_to_update, col] = value
    else:
        df = pd.concat([df, pd.DataFrame([extracted_data], columns=df.columns)], ignore_index=True)

    # Save the updated DataFrame to CSV
    df.to_csv(csv_file, index=False)
    
    print(f"✅ Recipe '{name_to_update}' added/updated successfully!")
    
    # Display the updated DataFrame
    return df

# Now, whenever you define a new recipe_text, call:
# add_recipe(recipe_text, patterns)
```
